```python
import pandas as pd
import requests
import itertools


url = 'https://fantasyfootballcalculator.com/draft/6541965'

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}
r = requests.get(url, headers=header)
df = pd.read_html(r.text, header = 1)
df = df[0]
df = df.iloc[:-1,1:13]

for i in range(0,len(list(df))):

     df.iloc[:,i] = df.iloc[:, i].str.replace(r'(.*[A-Z]{0,1})([A-Z][a-z])', r'\1  \2')
     df.iloc[:,i] = df.iloc[:, i].str.replace(r'(.*) ([A-Z]{2})(.*)', r'\1')

results_list = []

for i in range(0,len((df))):
    results_list.append(df.iloc[i,:])

results_list = list(itertools.chain.from_iterable(results_list))
```
