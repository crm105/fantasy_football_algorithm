```python
import pandas as pd
import requests
import itertools
import difflib
import numpy as np

ffc = pd.read_csv('ffc_names.csv')
results = pd.DataFrame()

url_list = ["6584965", '6541965', '6584960', '6584958', '6584803' , '6584801', '6584798', '6584796', '6584793', '6584642', '6584634', 
           '6584630', '6584627', '6584497', '6584495', '6584492', '6584490', '6584487', '6584487', '6584485', '6584483', '6584480', 
           '6584370', '6584368', '6584366', '6584361', '6584358', '6584351', '6584238']
url = '6541965'

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

for p in url_list:
    url = ('https://fantasyfootballcalculator.com/draft/' + p)
    r = requests.get(url, headers=header)
    df = pd.read_html(r.text, header = 1)
    df = df[0]
    df = df.iloc[:-1,1:13]

    for i in range(0,len(list(df))):

         df.iloc[:,i] = df.iloc[:, i].str.replace(r'(.*[A-Z]{0,1})([A-Z][a-z])', r'\1  \2')
         df.iloc[:,i] = df.iloc[:, i].str.replace(r'(.*) ([A-Z]{2})(.*)', r'\1')

    full_list = []
    results_list = []

    for i in range(0,len((df))):
        results_list.append(df.iloc[i,:])

    full_list.append ( list(itertools.chain.from_iterable(results_list)))
    try:
  

        results[p] = list(itertools.chain.from_iterable(results_list))
    except Exception:
        pass

results = results.reset_index()
```

```python
ffc = pd.read_csv('ffc_names.csv', names = ['Rank', 'name'])
ffc = ffc.drop('Rank', axis = 1)
rem_name = []
for i in range(0, len(ffc)):
    rem_name.append( difflib.get_close_matches(ffc.iloc[i,0], full_list[0], cutoff = .15)[0])

indexer = ffc.index
ffc.loc[indexer, "name"] = rem_name

results['pick'] = np.arange(len(results))
for i in range(1,len(list(results))-1):
    drafted = []
    names = []
    for j in ffc['name']:
        try:
            drafted.append(int (results[results.iloc[:,i] == j]['pick'].values))
            names.append(j)
        except:
            pass
    df = pd.DataFrame({i: drafted, 'name': names})
    df[i] = df[i] + 1
    df = df.drop_duplicates()
    ffc = ffc.merge(df, how = 'left')
    ffc = ffc.drop_duplicates()
ffc.head()

ffc.to_csv('mock_results.csv')
        
```
